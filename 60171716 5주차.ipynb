```python
import numpy as np
import matplotlib.pyplot as plt

def main() :
    Score = np.random.randint(100, size = (10, 4))
    Q1(Score)
    Q2(Score)
    Q3(Score)
    Q4(Score)

def Q1(s) :
    # Slice each data from Score
    Korean = s[:, :1]
    English = s[:, 1:2]

    # Use scatter function
    plt.scatter(English, Korean)

    # Label each axis
    plt.xlabel("English")
    plt.ylabel("Korean")

    plt.show()

def Q2(s) :
    # Slice each data from Score
    Korean = s[:, :1]
    English = s[:, 1:2]
    Math = s[:, 2:3]
    Science = s[:, 3:4]
    
    # Divide into 4 area by using subplot function
    plt.subplot(2, 2, 1)
    plt.hist(Korean)
    plt.xlabel("Korean")
    plt.ylabel("Count")
    plt.subplot(2, 2, 2)
    plt.hist(English)
    plt.xlabel("English")
    plt.ylabel("Count")
    plt.subplot(2, 2, 3)
    plt.hist(Math)
    plt.xlabel("Math")
    plt.ylabel("Count")
    plt.subplot(2, 2, 4)
    plt.hist(Science)
    plt.xlabel("Science")
    plt.ylabel("Count")

    # Adjust space of each subplot
    plt.tight_layout()

    plt.show()

def Q3(s) :
    # Change 2D array into 1D array to use where function
    Korean = s[:, :1].ravel()

    # Get number of each grade by using where function
    a = np.where(Korean >= 90)
    b = np.where((Korean >= 80) & (Korean < 90))
    c = np.where((Korean >= 70) & (Korean < 80))
    d = np.where(Korean < 70)
    
    # Set ratio and labels 
    ratio = [a[0].size, b[0].size, c[0].size, d[0].size]
    labels = ['A', 'B', 'C', 'D']

    # Use pie function
    plt.pie(ratio, labels = labels, autopct = '%.1f%%')

    plt.show()

def Q4(s) :
    # Change 2D array into 1D array to use boxplot function
    Korean = s[:, :1].ravel()
    English = s[:, 1:2].ravel()
    Math = s[:, 2:3].ravel()
    Science = s[:, 3:4].ravel()

    # Use boxplot function
    data = [Korean, English, Math, Science]
    plt.boxplot(data, labels = ['Korean', 'English', 'Math', 'Science'])

    plt.show()

main()
```
